### *This workbook focuses on clustering the neighborhoods in Toronto*

```python
#Importing required packages
import pandas as pd
import folium # map rendering library
```

```python
#reading the Toronto dataset with geospatial information
df=pd.read_csv("Toronto Neighborhood with Geospatial.csv")
df.head()
```

#### Exploring the Toronto neighborhoods- the one with Toronto in the Borough name

```python
#filetering neighborhoods with boroughs with Toronto name
df_neighborhood=df[df['Borough'].str.contains("Toronto")]
df_neighborhood.head()
```

*Exploring the dataset with Borough name containing "Toronto"*

```python
df_neighborhood['Borough'].value_counts()
```

*Create map to visualize the clusters in the Toronto neighborhood*

```python
map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to map
for postcode, lat, lng, borough, neighborhood in zip(df_neighborhood['PostCode'], df_neighborhood['Latitude'], df_neighborhood['Longitude'], df_neighborhood['Borough'], df_neighborhood['Neighborhood']):
    label = '{}, {}, {}'.format(neighborhood, borough,postcode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3185cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
